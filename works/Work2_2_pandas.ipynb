```python
# Відсортуйте таблицю за віком у порядку зростання
print("за віком у порядку зростання:")
workers.sort_values(by="Вік", ascending=True)
```

```python
# Відсортуйте таблицю за зарплатою в порядку спадання.
print("за зарплатою в порядку спадання:")
workers.sort_values(by="Зарплата", ascending=False)
```

<h2>Завдання 5</h2>
Є список студентів із зазначенням: "Ім'я", "Курс", "Оцінка".
<ul>
    <li>Створіть DataFrame і заповніть його не менш ніж вісьмома рядками.</li>
    <li>Виведіть вихідний DataFrame.</li>
    <li>Обчисліть середню оцінку за кожним курсом.</li>
    <li>Знайдіть максимальну та мінімальну оцінку за кожним курсом.</li>
</ul>

```python
students = pd.read_csv("./students02_2.csv")
students.head(3)
```

```python
# Обчисліть середню оцінку за кожним курсом.
# Групування - розділення дані за однаковими значеннями полів
marks = students.groupby('Курс')['Оцінка']
# marks  # <pandas.api.typing.SeriesGroupBy object at 0x00000172B55652B0>
print("Середні оцінки за курсами:", marks.mean(), sep='\n')
```

```python
# Знайдіть максимальну та мінімальну оцінку за кожним курсом.
print("Максимальні оцінки за курсами:", marks.max(), sep='\n')
print("Мінімальні оцінки за курсами:", marks.min(), sep='\n')
```
