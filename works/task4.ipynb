## Завдання 1
Є дані про планові та фактичні продажі за місяцями. Потрібно порівняти два ряди на одному графіку.
 * Створіть список місяців і два списки значень: "План" і "Факт".
 * Побудуйте один графік із двома лініями.
 * Додайте підписи осей і заголовок.
 * Додайте легенду для ліній.
 * Відобразіть графік на екрані.


```python
import matplotlib.pyplot as plt

months = ["Січень", "Лютий", "Березень", "Квітень", "Травень", "Червень", "Липень", "Серпень"]
plan = [120, 150, 140, 180, 200, 210, 190, 230]
fact = [115, 160, 135, 175, 210, 205, 200, 245]

plt.figure(figsize=(10, 5))

plt.plot(months, plan, label='План', marker='o', linestyle='--')
plt.plot(months, fact, label='Факт', marker='s', linestyle='-')

plt.title('Порівняння планових та фактичних продажів')
plt.xlabel('Місяць')
plt.ylabel('Обсяг продажів (грн)')

plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Завдання 2
Дано значення вікових груп 100 осіб. Потрібно показати розподіл.
 * Створіть список (або масив) зі 100 цілих значень віку.
 * Побудуйте гістограму розподілу.
 * Додайте вертикальну лінію середнього значення.
 * Підпишіть осі та додайте заголовок.
Відобразіть графік на екрані.


```python
import numpy as np

np.random.seed(42) 
ages = np.random.randint(10, 81, size=100)

mean_age = np.mean(ages)

plt.figure(figsize=(10, 6))

plt.hist(ages, bins=15, color='blue', edgecolor='black', alpha=0.7)

plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Середнє: {mean_age:.1f}')

plt.title('Розподіл віку групи зі 100 осіб')
plt.xlabel('Вік (роки)')
plt.ylabel('Кількість осіб')

plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

## Завдання 3
Є результати іспитів за трьома групами студентів. Потрібно порівняти розкид оцінок між групами.
 * Створіть три набори числових результатів (по одній вибірці на групу).
 * Побудуйте графік boxplot для трьох груп поруч.
 * Підпишіть групи по осі X.
 * Додайте підписи осей і заголовок.
 * Відобразіть графік на екрані.


```python
group_A = np.random.normal(75, 10, 30)
group_B = np.random.normal(82, 5, 30)
group_C = np.random.normal(65, 15, 30)
data = [group_A, group_B, group_C]
my_labels = ['Група А', 'Група Б', 'Група В']

plt.figure(figsize=(10, 6))

plt.boxplot(data, tick_labels=my_labels, patch_artist=True, showmeans=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red', linewidth=2))

plt.title('Порівняння результатів іспитів за групами')
plt.ylabel('Оцінки (бали)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Завдання 5
Є погодинні дані навантаження сервера за добу. Потрібно показати зміни і виділити область між кривою і віссю X.
 * Створіть список годин (0-23) і відповідні значення навантаження.
 * Побудуйте лінійний графік навантаження.
 * Зафарбуйте область під кривою.
 * Додайте сітку, підписи осей і заголовок.
 * Відобразіть графік на екрані.


```python
hours = list(range(24))
load = [15, 12, 10, 8, 9, 12, 25, 45, 60, 75, 80, 85, 82, 78, 75, 80, 90, 95, 85, 70, 55, 40, 25, 20]

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(hours, load, color='teal', linewidth=2, label='Server Load')
ax.fill_between(hours, load, color='teal', alpha=0.3) 

ax.set_title('Hourly Server Load (24-hour period)')
ax.set_xlabel('Hour of Day')
ax.set_ylabel('Load (%)')

ax.set_xticks(np.arange(0, 24, 2))

ax.grid(True, linestyle=':', alpha=0.6)
ax.legend()

fig.tight_layout()
plt.show()
```

## Завдання 6
Дано чотири набори метрик продукту: конверсія, утримання, середній чек, кількість замовлень. Потрібно показати їх на окремих панелях.
 * Створіть чотири набори числових значень за однією шкалою часу.
 * Створіть сітку з чотирьох підграфіків 2×2.
 * На кожному підграфіку побудуйте відповідний графік.
 * Додайте заголовки для кожного підграфіка та загальний заголовок.
 * Відобразіть результат на екрані.


```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']
conversion = [2.5, 2.7, 3.1, 2.9, 3.5, 4.0, 4.2]
retention = [45, 47, 46, 50, 52, 55, 58]
avg_check = [120, 125, 122, 130, 145, 140, 150]
orders = [500, 550, 600, 580, 700, 750, 800]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(months, conversion, color='green', marker='o')
axs[0, 0].set_title('Conversion Rate (%)')
axs[0, 0].grid(True, alpha=0.3)

axs[0, 1].plot(months, retention, color='blue', marker='s')
axs[0, 1].set_title('Retention Rate (%)')
axs[0, 1].grid(True, alpha=0.3)

axs[1, 0].plot(months, avg_check, color='orange', marker='^')
axs[1, 0].set_title('Average Check ($)')
axs[1, 0].grid(True, alpha=0.3)

axs[1, 1].plot(months, orders, color='purple', marker='v')
axs[1, 1].set_title('Total Orders')
axs[1, 1].grid(True, alpha=0.3)

fig.suptitle('Product Performance Metrics (2026)', fontsize=16)

fig.tight_layout(rect=[0, 0.03, 1, 0.95]) 

plt.show()
```
