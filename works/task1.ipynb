## Завдання 1

Вам потрібно порахувати ключові показники за даними про температуру. Деталі реалізації:
    
* Створіть Notebook homework_temperature.ipynb.
* У кодовій комірці створіть список температур: [2, 5, 4, 6, 7, 3, 1].
* За допомогою функцій Python обчисліть:
    - середнє значення;
    - максимальне і мінімальне значення;
    - різницю між максимальним і мінімальним значенням.
* Оформіть результати у вигляді таблиці (Markdown).

```python
from IPython.display import Markdown

temperatures = [2, 5, 4, 6, 7, 3, 1]

avg_temp = sum(temperatures) / len(temperatures)
max_temp = max(temperatures)
min_temp = min(temperatures)
diff = max_temp - min_temp

table_content = f"""
| Показник | Значення |
| :--- | :--- |
| **Середнє** | {avg_temp:.2f} |
| **Максимальне** | {max_temp:.2f} |
| **Мінімальне** | {min_temp:.2f} |
| **Різниця між макс. і мін.** | {diff:.2f} |
"""

Markdown(table_content)
```

## Завдання 2

Вам необхідно розрахувати показники за списком щоденних витрат. Деталі реалізації:

* Створіть Notebook homework_expenses.ipynb.
* У кодовій комірці створіть список: [1200, 1500, 800, 600, 1000, 400, 900].
* Розрахуйте:
    - загальну суму витрат;
    - середнє значення;
    - максимальна і мінімальна витрата;
    - кількість днів, коли витрата була вищою за 1000.
* Відобразіть усі результати в таблиці

```python
from IPython.display import Markdown

expenses = [1200, 1500, 800, 600, 1000, 400, 900]
sum = sum(expenses)
avg = sum / len(expenses)
max = max(expenses)
min = min(expenses)
days_above_1000 = len([x for x in expenses if x > 1000])

table = """
| Показник | Значення |
| :--- | :--- |
| Сума | %.2f |
| Середнє | %.2f |
| Максимальне | %.2f |
| Мінімальне | %.2f |
| Кількість днів > 1000 | %d |
""" % (sum, avg, max, min, days_above_1000) 
Markdown(table)
```

## Завдання 3

Вам потрібно обробити список голосів і автоматично знайти лідера. Деталі реалізації:

* Створіть Notebook homework_votes.ipynb.
* У кодовій комірці створіть список:
votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
* Використовуючи вбудовані засоби Python (наприклад, count(), dict, або collections.Counter), порахуйте, скільки разів зустрічається кожен фрукт.
* Напишіть код для визначення фрукта-лідера (зустрічається найбільшу кількість разів).
* Оформіть результати в таблиці: фрукт → кількість.
* Окремо покажіть топ-3.

```python
from IPython.display import Markdown

votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
fruits_count = {fruit: votes.count(fruit) for fruit in set(votes)}

sorted_votes = sorted(fruits_count.items(), key=lambda x: x[1], reverse=True)
leader = sorted_votes[0][0]
top3 = sorted_votes[:3]

table_rows = "\n".join([f"| {fruit} | {count} |" for fruit, count in fruits_count.items()])

table = f"""
### Результати голосування
| Фрукт | Кількість |
| :--- | :--- |
{table_rows}

**Фрукт-лідер:** {leader}

**Топ-3:** {', '.join([f"{f} ({c})" for f, c in top3])}
"""

Markdown(table)
```

## Завдання 4

Порівняйте результати з предметів і обчисліть основні показники. Деталі реалізації:

* Створіть Notebook homework_scores.ipynb.
* У кодовій комірці створіть списки:
  math = [80, 92, 75, 60, 88]
  physics = [70, 65, 90, 85, 78]
  english = [95, 100, 85, 90, 87]
* Для кожного предмета обчисліть:
    - середній бал;
    - мінімальний бал;
    - максимальний бал.
* Додатково знайдіть предмет з найвищим середнім результатом (кодом, без тексту).
* Представте результати в таблиці.

```python
from IPython.display import Markdown

# 1. Списки балів
math_scores = [80, 92, 75, 60, 88]
physics_scores = [70, 65, 90, 85, 78]
english_scores = [95, 100, 85, 90, 87]

m_avg, m_min, m_max = sum(math_scores)/len(math_scores), min(math_scores), max(math_scores)
p_avg, p_min, p_max = sum(physics_scores)/len(physics_scores), min(physics_scores), max(physics_scores)
e_avg, e_min, e_max = sum(english_scores)/len(english_scores), min(english_scores), max(english_scores)

subjects_dict = {"Математика": m_avg, "Фізика": p_avg, "Англійська": e_avg}
best_subject = max(subjects_dict, key=subjects_dict.get)

table = f"""
| Предмет | Сер. бал | Мін. бал | Макс. бал |
| :--- | :---: | :---: | :---: |
| Математика | {m_avg:.2f} | {m_min} | {m_max} |
| Фізика | {p_avg:.2f} | {p_min} | {p_max} |
| Англійська | {e_avg:.2f} | {e_min} | {e_max} |

**Предмет з найвищим середнім результатом:** {best_subject}
"""

Markdown(table)
```
