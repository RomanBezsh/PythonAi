## Завдання 1
У вас є список чисел від 1 до 10 та їхні квадрати. Потрібно побудувати графік залежності квадрата числа від самого числа.
- Створіть два списки: перший містить числа від 1 до 10, другий — їхні квадрати.
- Побудуйте лінійний графік, де по осі X відкладаються числа, а по осі Y — їхні квадрати.
- Додайте заголовок "Графік функції y = x²" і підписи осей "X" і "Y".
- Зробіть лінію червоного кольору та позначте кожну точку маркером.
- Відобразіть графік на екрані.

```python
import matplotlib.pyplot as plt  # pip install matplotlib
```

```python
# Створіть два списки: перший містить числа від 1 до 10, другий — їхні квадрати.
x = [i for i in range(1,11)]
y = [i ** 2 for i in x]
y
```

```python
fig, ax = plt.subplots()
# Побудуйте лінійний графік, де по осі X відкладаються числа, а по осі Y — їхні квадрати.
# Зробіть лінію червоного кольору та позначте кожну точку маркером.
ax.plot(x, y, 'o:r')  # marker-style-color 's--g' - square marker, dashed line, green color 
# Додайте заголовок "Графік функції y = x²" і підписи осей "X" і "Y".
ax.set_title("Графік функції y = x²")  
ax.set_xlabel("THE X")
ax.set_ylabel("THE Y")
plt.show()
```

## Завдання 2
Є дані про продажі за 6 місяців: [150, 200, 250, 300, 280, 320]. Потрібно показати, як змінювалися продажі за місяцями.
- Створіть список місяців і список продажів.
- Побудуйте стовпчасту діаграму: по осі X — місяці, по осі Y — продажі.
- Додайте підписи осей і заголовок діаграми.
- Зробіть стовпці синього кольору.
- Відобразіть діаграму на екрані.

```python
# Створіть список місяців і список продажів.
sales = [150, 200, 250, 300, 280, 320]
monthes = ["Січ","Лют","Бер","Кві","Тра","Чер"]
```

```python
# Побудуйте стовпчасту діаграму: по осі X — місяці, по осі Y — продажі.
fig, ax = plt.subplots()  # починаємо новий графік (діаграму)
ax.bar(monthes, sales, color="blue")
ax.set_title("Продажі за пів року")  
ax.set_xlabel("Місяць")
ax.set_ylabel("Продажі, шт")
plt.show()
```

## Завдання 3
Дано три категорії товарів: "Продукти", "Одяг", "Техніка" та їхні частки продажів [40, 30, 30]. Потрібно показати розподіл продажів між категоріями.
- Створіть список категорій і список їхніх часток.
- Побудуйте кругову діаграму.
- Додайте підписи до кожного сектора.
- Покажіть процентні значення всередині кругової діаграми.
- Відобразіть результат.

```python
def to_percent(n) :
    return f"{n:.1f} %"

categories  = ["Продукти", "Одяг", "Техніка"]
sales_parts = [40, 30, 40]
fig, ax = plt.subplots()
ax.pie(sales_parts,          # Кругова діаграма - показує частки категорій у складі цілого
       labels=categories,    # labels - підписи за діаграмою (перелік рядків)      
       autopct=to_percent)   # autopct - функція (лямбда) перетворення проценту до надпису
plt.show()
```

# Конспект
Графік-лінія будується тоді, коли гарантується, що між відомими точками лінія відповідає також за правильні значення. Зазвичай, це робиться для математичних формул, а також для неперервних даних (сейсмографи, тахографи тощо). Лініями не можна сполучати точки, які є результатами вимірювань (з похибкою) або одним з можливих, не закономірних значень (вага-зріст)

Інтерполяція - розрахунок значень у точках, що знаходяться між (інтер) точками з відомими значеннями.
В цій задачі точки сполучаються, але не завжди прямою лінією

Екстраполяція - розрахунок значень поза межами проведених вимірювань чи області визначених даних

Апроксимація - проведення лінії серед точок, значення яких не є гарантованим. Часто при цьому відома формула лінії.
