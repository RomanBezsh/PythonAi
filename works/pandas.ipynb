## Завдання 2
У файлі CSV збережено дані про фільми: "Назва", "Жанр", "Рік", "Рейтинг".
 * Прочитайте дані з CSV у DataFrame.

 * Визначте загальну кількість рядків і стовпців.
 * Визначте типи даних для кожного стовпця.
 * Перевірте наявність пропусків.


```python
movies = pd.read_csv("./notes02_2.csv")
movies.head()
```

```python
print("gereral count rows: %d and columns: %d" % (
    len(movies.index),
    len(movies.columns)
))
```

```python
print("типи даних для кожного стовпця:", movies.dtypes, sep='\n')
```

```python
# Перевірте наявність пропусків.
import numpy as np
print("------- Технічні деталі ---------")
empties = movies.isnull()
print(empties)
positions = np.where(empties)
print(positions)  # (array([6, 9]), array([1, 2]))
print(*zip(*positions))  # (np.int64(6), np.int64(1)) (np.int64(9), np.int64(2))
print("------- Підсумки ---------")
print("Пропуски знаходяться за індексами: ", 
      ", ".join(f"({x};{y})" for (x, y) in zip(*positions))
)
```

## Завдання 5
Є список студентів із зазначенням: "Ім'я", "Курс", "Оцінка".
 * Створіть DataFrame і заповніть його не менш ніж вісьмома рядками.
 * Виведіть вихідний DataFrame.
 * Обчисліть середню оцінку за кожним курсом.
 * Знайдіть максимальну та мінімальну оцінку за кожним курсом.


```python
students = pd.read_csv('./students02_2.csv')
students.head(3)
```

```python
marks = students.groupby('Курс')['Оцінка']
print("Середня оцінка за курсом", marks.mean())
```

```python
print("Максимальні оцінки:", marks.max(), sep='\n')
print("\nМінімальні оцінки:", marks.min(), sep='\n')
```
