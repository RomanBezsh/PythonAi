## Завдання 1
* Створіть масив із 20 цілих чисел у діапазоні від 0 до 50.
* Виведіть вихідний масив.
* Запитайте у користувача число (порогове значення).
* Визначте кількість елементів масиву, що перевищують введене користувачем значення, і виведіть результат.
* Знайдіть максимальне значення в масиві та позицію його першої появи, виведіть обидва результати.
* Відсортуйте масив за спаданням і виведіть відсортований масив.

```python
import numpy as np
```

```python
arr = np.random.randint(0, 50, size=20) 
arr
```

```python
threshold = int(input("Введіть порогове значення: "))
count_greater = np.sum(arr > threshold)
print(f"Кількість елементів, більших за {threshold}: {count_greater}")
```

```python
max_val = np.max(arr)
max_index = np.argmax(arr)
print(f"Максимальне значення: {max_val}, позиція першої появи: {max_index}")    
```

```python
sorted_arr = np.sort(arr)[::-1]
print(f"Відсортований масив за спаданням: {sorted_arr}")
```

## Завдання 2
* Запитайте в користувача верхню та нижню межу діапазону чисел.
* Створіть квадратну матрицю 5×5 з цілими значеннями в цьому діапазоні.
* Виведіть вихідну матрицю.
* Отримайте головну діагональ і виведіть діагональні елементи.
* Підрахуйте суму елементів головної діагоналі та виведіть результат.
* Обнуліть елементи строго вище головної діагоналі та виведіть змінену матрицю.

```python
import numpy as np
```

```python
low = int(input("Введіть нижню межу діапазону: "))
high = int(input("Введіть верхню межу діапазону: "))
```

```python
if low >= high:
    print("Помилка: нижня межа має бути меншою за верхню.")
else:
        matrix = np.random.randint(low, high + 1, size=(5, 5))
        print("\nВихідна матриця:\n", matrix)

        diagonal_elements = []
        for i in range(5):
            diagonal_elements.append(matrix[i, i])
        
        print(f"\nЕлементи головної діагоналі: {diagonal_elements}")

        diag_sum = sum(diagonal_elements)
        print(f"Сума елементів головної діагоналі: {diag_sum}")

        for i in range(5):       
            for j in range(5):    
                if j > i:         
                    matrix[i, j] = 0

        print("\nМатриця після обнулення (вище діагоналі):\n", matrix)
```
