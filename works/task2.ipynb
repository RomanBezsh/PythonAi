## Завдання 2
 * Запитайте в користувача верхню та нижню межу діапазону чисел.
 * Створіть квадратну матрицю 5×5 з цілими значеннями в цьому діапазоні.
 * Виведіть вихідну матрицю.
 * Отримайте головну діагональ і виведіть діагональні елементи.
 * Підрахуйте суму елементів головної діагоналі та виведіть результат.
 * Обнуліть елементи строго вище головної діагоналі та виведіть змінену матрицю.

```python
import numpy as np
```

```python
low = int(input("Введіть нижню межу діапазону: "))
high = int(input("Введіть верхню межу діапазону: "))
```

```python
if low >= high:
    print("Помилка: нижня межа має бути меншою за верхню.")
else:
        matrix = np.random.randint(low, high + 1, size=(5, 5))
        print("\nВихідна матриця:\n", matrix)

        diagonal_elements = []
        for i in range(5):
            diagonal_elements.append(matrix[i, i])
        
        print(f"\nЕлементи головної діагоналі: {diagonal_elements}")

        diag_sum = sum(diagonal_elements)
        print(f"Сума елементів головної діагоналі: {diag_sum}")

        for i in range(5):       
            for j in range(5):    
                if j > i:         
                    matrix[i, j] = 0

        print("\nМатриця після обнулення (вище діагоналі):\n", matrix)
```

 * Запитайте у користувача розмірність вихідної матриці (наприклад, 4×6).
 * Створіть матрицю з послідовно зростаючих цілих чисел заданої розмірності.
 * Виведіть вихідну матрицю.
 * Запитайте в користувача нові розміри (наприклад, 2×12) і перетворіть матрицю в зазначену форму.
 * Виведіть отриману матрицю.
 * Визначте мінімальні та максимальні значення в кожному рядку нової матриці та виведіть обидва одновимірні масиви.
 * Підрахуйте загальну суму всіх елементів нової матриці та виведіть результат.

```python
rows = int(input("Введіть довжину рядків: "))
cols = int(input("Введіть довжину стовбців: "))
```

```python
matrix = np.arange(rows * cols).reshape(rows, cols)
matrix
```

```python
new_rows = int(input("Введіть нову довжину рядків: "))
new_cols = int(input("Введіть нову довжину стовбців: "))
```

```python
resh = np.reshape(matrix, shape=(new_rows,new_cols))
print("Зміненна матриця за формою:\n",resh)
```

```python
row_min = np.min(resh, axis=1)
row_max = np.max(resh, axis=1)
print("Мінімальні значення в кожному рядку:", row_min)
print("Максимальні значення в кожному рядку:", row_max)
```

```python
total_sum = np.sum(resh)
print("Загальна сума всіх елементів матриці:", total_sum)
```
